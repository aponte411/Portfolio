Now, lets say I want to build a (hypothetical) basic nutritional index for people who want to eat high-protein, low-fat foods. I'll be using the protein and lipid column, and the following formula to 'score' each food in our data set:
 - score = (2 x protein) - (0.75 x lipids)

```python
weighted_protein = food_info['Protein_(g)']* 2
weighted_fat = food_info['Lipid_Tot_(g)'] * 0.75
initial_rating = weighted_protein - weighted_fat
print(initial_rating)
```

Right now, my data isn't normalized. The range of values varies greatly - e.g kilo-calories, mg, g. etc. So I want to first rescale my data using the rescaling formula:
 - rescaled_column = (original_column - min_value_in_column) / (max_value_in_column - min_value_in_column)

```python
# creating rescaling function
def rescaling(column_name):
    rescaled = (food_info[column_name] - food_info[column_name].min()) / (food_info[column_name].max() - food_info[column_name].min())
    return rescaled
```

```python
normalized_protein = rescaling('Protein_(g)')
normalized_lipids = rescaling('Lipid_Tot_(g)')
normalized_water = rescaling('Water_(g)')
normalized_carbs = rescaling('Carbohydrt_(g)')
```

```python
# adding the normalized columns to the dataframe
food_info['Normalized_Protein'] = normalized_protein
food_info['Normalized_Fat'] = normalized_lipids
food_info['Normalized_Carbs'] = normalized_carbs
```

```python
food_info.head(1)
```

Now, im going to use the normalized protein and normalized fat columns to redo my nutritional index calculation. I will store those calculations as a new column in the DataFrame - 'Norm_Nutr_Index'.

```python
food_info['Norm_Nutr_Index'] = (2 * food_info['Normalized_Protein']) - (0.75 * food_info['Normalized_Fat'])
```
